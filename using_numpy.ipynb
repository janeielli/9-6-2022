```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([1,2,3,4]) # 1-D array
print("Array dimension: ",a.shape)
print("Number of elements in the array: ",a.size) 

b = np.array([[3,4,5,3],[7,1,2,9],[10,2,3,7]]) # 2-D array
print("Array dimension: ",b.shape)
print("Number of elements in the array: ",b.size) 
```

```python
#initializing np array
c = np.zeros((3,4))
print(c)

d = np.ones((3,4))
print(d)

e = np.eye(4)
print(e)
```

```python
#Accessing arrays elements
print("Array: \n", b)
print("element: b[{},{}]={}".format(1,2,b[0,1]))
```

```python
# Deleting an element in an Array
print("Original matrix: \n", b)
d = np.delete(b,1,0)
f = np.delete(b,1,1)
print("Deleting the second row: \n", d)
print("Deleing the second column: \n", f)
```

```python
b[0,:] = [2,4,5,6]
print(b) 
```

```python
#Convert 2-D array into a 1-D array
t = np.ndarray.flatten(b)
print(t)
```

```python
#Finding values given a condition
print(t>5)
print("\nVerify b[{},{}] = {} ----> {}".format(0,0,b[0,0],(b[0,0]>5)))
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'b*')
plt.axis([0,6,0,20])
```

```python
# Creating an array with values from 0 to 5 using 0.2 as step
t = np.arange(0.,5.,0.2) 
# Plotting 3 different functions in one figure: (t,t,'r-'), (t,t**2,'bs'), (t,t**3,'g^')
plt.plot(t,t,'r-', t,t**2,'bs', t,t**3,'g^')
plt.show()
```
